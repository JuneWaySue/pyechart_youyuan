```python
import pymysql
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyecharts.charts import Bar,Pie,Page,Geo,Line
from pyecharts.globals import ChartType
import pyecharts.options as opts
```

```python
conn=pymysql.connect(host='localhost',user='',password='',database='db',charset='utf8')
cursor=conn.cursor()
sql='select * from youyuan'
cursor.execute(sql)
items=cursor.fetchall()
cursor.close()
conn.close()
```

```python
name=[]
address=[]
age=[]
height=[]
salary=[]
house=[]
hobby=[]
image=[]
motto=[]
detail=[]
boy_condition=[]
url=[]

for item in items:
    name.append(item[1])
    address.append(item[2])
    age.append(item[3])
    height.append(item[4])
    salary.append(item[5])
    house.append(item[6])
    hobby.append(json.loads(item[7]))
    image.append(json.loads(item[8]))
    motto.append(item[9])
    detail.append(json.loads(item[10]))
    boy_condition.append(json.loads(item[11]))
    url.append(item[12])
df=pd.DataFrame([name,address,age,height,salary,house,hobby,image,motto,detail,boy_condition,url]).T
df.columns=['昵称','地址','年龄','身高','工资','住房','爱好','相册','宣言','资料','择偶','空间']
df
```
